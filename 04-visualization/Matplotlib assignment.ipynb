Charts to plot:

```python
url=r'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
num = titanic.sex.value_counts()
lab = num.index
```

#### 1. Create a pie chart presenting the male/female proportion

```python
plt.axis("equal")
plt.pie(num, labels=lab, autopct='%1.1f%%',radius=1.5, shadow=True)
plt.show()
```

#### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure(figsize=(20, 15))
plt.xlabel('Age')
plt.ylabel("Fare paid")
plt.scatter(titanic['age'], titanic['fare'], c=titanic['sex'].map({'male': 1, 'female': 2}))
plt.title('Fare vs Age Chart')
```
